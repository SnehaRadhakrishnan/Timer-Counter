```python
#import module time
import time

#declare class and function
class CountdownTimer:
    def __init__(self, duration):
        self.duration = duration
        self.start_time = None
        self.paused = False
        self.paused_duration = 0
  #to start the timer  
    def start(self):
        self.start_time = time.time()
        while True:
            if self.paused:
                time.sleep(1)
                continue
            elapsed_time = time.time() - self.start_time - self.paused_duration
            remaining_time = self.duration - elapsed_time
            if remaining_time <= 0:
                print("And your time is up!!!!!")
                break
            print(f"Remaining time: {int(remaining_time)} seconds")
            time.sleep(1)

#to reset the timer   
    def reset(self):
        self.start_time = None
        self.paused = False
        self.paused_duration = 0

#to pause the timer      
    def pause(self):
        if self.start_time is not None and not self.paused:
            self.paused = True
            self.pause_start_time = time.time()
            print("Timer paused.")
            
#to resume the timer    
    def resume(self):
        if self.start_time is not None and self.paused:
            self.paused = False
            self.paused_duration += time.time() - self.pause_start_time
            print("Timer resumed.")

#take the user input and loop until exit
def main():
    duration = int(input("Enter the countdown duration in seconds: "))
    timer = CountdownTimer(duration)
 #given choice to input and loop until exit  
    while True:
        command = input("Enter a command \n1.Start\n2.Pause\n3.Resume\n4.Reset\n5.Stop")
        if command == "1":
            timer.start()
        elif command == "2":
            timer.pause()
        elif command == "3":
            timer.resume()
        elif command == "4":
            timer.reset()
            print("Timer reset.")
        elif command == "5":
            break
        else:
            print("Invalid command.")

# to store code that should only run when your file is executed as a script.
if __name__ == "__main__":
    main()

    
```
